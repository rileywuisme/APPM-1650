```python
import math
import random
import numpy as np
import matplotlib.pyplot as plt
```

Plot $ y = x + 4 \sin(x) $ on $[-2\pi, 2\pi]$

```python
xvals = np.linspace(-2*math.pi, 2*math.pi, 100)
yvals = [x + 4*math.sin(x) for x in xvals]

plt.figure(figsize=(3.5, 2.5))
plt.plot(xvals, yvals, lw = 2, alpha=0.6)
plt.ylim(-6, 6)
plt.title('Sine Curve', size = 18)
plt.ylabel('sine values')
plt.xlabel('x values')
plt.xticks([-2*math.pi + i*math.pi for i in range(5)], 
          ['-2pi', 'pi', 0, 'pi', '2pi']) #list comprehension that produces [-2pi, -pi, 0, pi , 2pi]
plt.grid()
plt.axhline(0, c = 'k')
plt.show()
```

### Box Plot  
List of 9 random integers drawn from [1, 20]

```python
randnums = [random.randint(1, 20) for i in range(9)]
randnums, sorted(randnums)
```

```python
plt.boxplot(randnums)
plt.figure(figsize = (3, 2))
plt.show()
```

```python
plt.boxplot(randnums, vert = False)
plt.show()
```

### Subplots

```python
plt.figure(figsize = (5, 5))

plt.subplot(221) #num rows, num cols, which plot
plt.plot(randnums)
plt.title('line')

plt.subplot(2, 2, 2) #num rows, num cols, which plot
plt.plot(randnums, 'o')
plt.title('scatter')

plt.subplot(2, 2, 3) #num rows, num cols, which plot
plt.bar(range(9), randnums)
plt.title('bar')

plt.subplot(2, 2, 4) #num rows, num cols, which plot
plt.boxplot(randnums)
plt.title('box')

plt.show()
```

### Specifying Colors

```python
nums = list(range(25, 0, -1))

for i, n in enumerate(nums):
    plt.bar(i, n)

plt.show()
```

Suppose there are 2 colors. How do we alternate the bar colors?

```python
colors = 'br'

nums = list(range(25, 0, -1))

for i, n in enumerate(nums):
    plt.bar(i, n, color = colors[i % 2])

plt.show()
```

### Displaying Text

```python
plt.text(0.2, 0.4, 'alice', size=18)
plt.text(0.4, 0.6, 'bob', size=24, c='m', rotation=90) #counterclockwise rotation in degrees
plt.show()
```

Display 200 'CUB' in random x, y locations with random rotations in different colors. 

```python
xval = [random.random() for n in range(200)]
yval = [random.random() for i in range(200)]
for n in range(200):
    plt.text(xval[n], yval[n], 'CUB')
```

```python
word = 'CUB'
plt.axis('off')

for n in range(200):
    plt.text(random.random(), random.random(), word, rotation=random.randint(0, 360), color='C' + str(n % 10))
```
