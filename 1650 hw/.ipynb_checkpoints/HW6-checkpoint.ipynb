## Sine and Cosine Curves

The following graph shows the function $ y = 2x\sin(x)$, $ y = x \cos(x)$, and $ y = 2 \sin^2(x) - 6 \cos^2(x)$ on $[-2\pi, 2\pi]$

```python
xvals = np.linspace(-2*math.pi, 2*math.pi, 100)
yval1 = [2*x*math.sin(x) for x in xvals]
yval2 = [x*math.cos(x) for x in xvals]
yval3 = [2*(math.sin(x))**2 - 6*((math.cos(x))**2) for x in xvals]

plt.figure(figsize =(7, 4.5))
plt.plot(xvals, yval1)
plt.plot(xvals, yval2)
plt.plot(xvals, yval3)
plt.ylim(-10, 6)

plt.show()
```

## Frequency of Rock, Paper, Scissors

The following bar graph shows the frequency of rock, paper, and scissors from a randomly generated list. 

```python
def rps_freq(nsim):
    result = [random.choice(['Rock', 'Paper', 'Scissors']) for n in range(nsim)]
    for r in result:
        res = [result.count('Rock'), result.count('Paper'), result.count('Scissors')]
        plt.bar(range(3), res, color='b')

    return res
```

```python
rps_freq(100)
```

## Normal Distribution

The following graph shows the graph when $\sigma = 0.5, 1.0, 1.5$

```python
def normal(x, sigma):
    first = 1 / (sigma * math.sqrt(2*math.pi))
    second = math.e ** (-x**2 / (2*sigma**2))
    return first * second

plt.figure(figsize=(5, 3))
xvals = np.linspace(-4, 4, 100)
plt.plot(xvals, normal(xvals, 0.5))
plt.plot(xvals, normal(xvals, 1.0))
plt.plot(xvals, normal(xvals, 1.5))

plt.show()
```

The following graph shows 30 different values of $\sigma$ on $[-4, 1]$

```python
for sig in range(5, 36):
    plt.plot(xvals, normal(xvals, sig/5))
```

The following is a boxplot of the same result. 

```python
norm = [normal(random.uniform(-4, 4), 1) for n in range(20)]

plt.boxplot(norm)
plt.show()
```

## V-Shaped Lines

```python
xvals = np.linspace(-8, 8, 100)

for n in range(15):
    yvals = [abs(x) + n/2 for x in xvals]
    plt.plot(xvals, yvals, 'C'+str(6 + n % 5))

plt.axis('equal')

for n in range(15):
    yvals = [-abs(x) +n/2 +8 for x in xvals]
    plt.plot(xvals, yvals, 'C'+str(5 + n % 5))

plt.axis('off')
```

## Circular Sectors

The following function `circ_sectors(n)` will output a circle with `n` equal sectors. 

```python
def circ_sectors(n):
    xvals = np.linspace(-1, 1, 100)
    yvals_up = [math.sqrt(1 - x**2) for x in xvals]
    yvals_bot = [-math.sqrt(1 - x**2) for x in xvals]
    plt.figure(figsize=(3, 3))
    plt.plot(xvals, yvals_up, 'b')
    plt.plot(xvals, yvals_bot, 'b')
    angle = np.linspace(0, 2*math.pi, n+1)
    for i in range(n):
        plt.plot([0, math.sin(angle[i])], [0, math.cos(angle[i])])
    plt.axis('equal')
    plt.show()
```

```python
circ_sectors(7)
```

```python
circ_sectors(40)
```

## Star

The following function `star(n)` will return a star with `n` spikes!

```python
def star(n):
    xvals = np.linspace(-1, 1, 100)
    yvals_up = [math.sqrt(1 - x**2) for x in xvals]
    yvals_down = [- math.sqrt(1 - x**2) for x in xvals]
    plt.plot(xvals, yvals_up)
    plt.plot(xvals, yvals_down, 'C0')
    for i in range(10):
        x = [math.cos(i*(2*math.pi / n)-(math.pi/n)), 10*math.cos(i*(2*math.pi/n)), math.cos(i*(2*math.pi/n)+(math.pi/n))]
        y = [math.sin(i*(2*math.pi / n)-(math.pi/n)), 10*math.sin(i*(2*math.pi/n)), math.sin(i*(2*math.pi/n)+(math.pi/n))]
        plt.plot(x, y)

    plt.axis('equal')
    plt.axis('off')
    plt.show()
```

```python
star(10)
```
