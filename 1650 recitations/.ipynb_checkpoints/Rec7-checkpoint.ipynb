### Graphing Functions

**Plot the functions** $f(x) = x\cos x$, $g(x) = -f(x)$, and $h(x) = x\sin x$ on $[-2\pi, 2\pi]$.

* **Vary the line styles**.
* **Add a `legend`** by including the `label=` parameter in the calls to `plot()`.

<img src="http://www.coloradomath.org/python/xcosx-graphs-legend.jpg" width="432" height="288" />

* **Vary the thickness of the lines and change the colors** so that $f$ is magenta ('m'), $g$ is cyan ('c'), and $h$ is a transparent blue ('b').

  * Use the parameter `lw=` to change the thickness of lines. The default value is 1.
  * Use the parameter `alpha=` to change the transparency of lines from 0 (transparent) to 1 (opaque), the default.

  <img src="http://www.coloradomath.org/python/xcosx-graphs-alpha.jpg" width="432" height="288" />

````python
xvals = np.linspace(-2*math.pi, 2*math.pi, 200)
f_yvals = [x * math.cos(x) for x in xvals]
g_yvals = [ - x * math.cos(x) for x in xvals]
h_yvals = [x * math.sin(x) for x in xvals]

plt.plot(xvals, f_yvals,'.', ms=1)
plt.plot(xvals, g_yvals, '--')
plt.plot(xvals, h_yvals, '-.')

plt.legend(['x*cos(x)', '-x*cos(x)', 'x*sin(x)'])

plt.show()
````

````python
xvals = np.linspace(-2*math.pi, 2*math.pi, 200)
f_yvals = [x * math.cos(x) for x in xvals]
g_yvals = [ - x * math.cos(x) for x in xvals]
h_yvals = [x * math.sin(x) for x in xvals]

plt.plot(xvals, f_yvals, 'm', lw=2, alpha=0.8)
plt.plot(xvals, g_yvals, 'c', lw=6, alpha=0.5)
plt.plot(xvals, h_yvals, 'b', lw=3, alpha=0.5)

plt.legend(['x*cos(x)', '-x*cos(x)', 'x*sin(x)'])

plt.show()
````

### Creating Patterns
Plot 10 line graphs of $y=x\cos x$, **varying the vertical offsets**.

<img src="http://www.coloradomath.org/python/xcosx-vert.jpg" width="432" height="288" />

````python
xvals = np.linspace(-2*math.pi, 2*math.pi, 100)
yvals = [x*math.cos(x) for x in xvals]

for n in range(10):
    yvals = [x*math.cos(x) + n for x in xvals]
    plt.plot(xvals, yvals)

plt.show()
````

Now produce the same graph, cycling through the **three default Python colors** `'C0'`, `'C1'`, and `'C2'`. (*Hint:* Use `%`.)

<img src="http://www.coloradomath.org/python/xcosx-vert-cycle.jpg" width="432" height="288" />

````python
xvals = np.linspace(-2*math.pi, 2*math.pi, 100)
yvals = [x*math.cos(x) for x in xvals]

for n in range(10):
    yvals = [x*math.cos(x) + n for x in xvals]
    plt.plot(xvals, yvals, 'C'+str(n%3))

plt.show()
````

Now **vary the horizontal offsets**.

<img src="http://www.coloradomath.org/python/xcosx-horiz.jpg" width="432" height="288" />

````python
xvals = np.linspace(-2*math.pi, 2*math.pi, 100)
yvals = [x*math.cos(x) for x in xvals]

for n in range(10):
    plt.plot(xvals + (n/5), yvals)

plt.show()
````

### Bar Charts and Box Plots

Generate **50 random real numbers** from the interval $[0, 10]$. Generate **another 25 random real numbers** from the interval $[10, 50]$. Combine the numbers into a single list.

* Create a vertical bar chart showing the distribution of random numbers.
* Include a `title()`, `xlabel()`, and `ylabel()`.
* Add horizontal gridlines using `grid(axis='y')`.
* Draw a horizontal orange line at $y=10$ using `axhline()`.
* Change the size of the display using `figure(figsize=())`.

<img src="http://www.coloradomath.org/python/random75-bar.jpg" width="864" height="576" />

* Use a horizontal bar chart to show the distribution.

<img src="http://www.coloradomath.org/python/random75-barh.jpg" width="432" height="288" />

* Make a boxplot. (Your plot may not have any outliers.)

<img src="http://www.coloradomath.org/python/random75-box.jpg" width="432" height="288" />


````python
random1 = [random.uniform(0, 10) for n in range(50)]
random2 = [random.uniform(10, 50) for n in range(25)]
com_random = random1 + random2

plt.figure(figsize=(8, 4))
plt.bar(range(75), com_random)
plt.title('Random Numbers', size = 15)
plt.xlabel('75 numbers')
plt.ylabel('random values')
plt.grid(axis='y')
plt.axhline(y=10, c='y')

plt.show()
````

````python
plt.figure(figsize=(5, 4))
plt.barh(range(75), com_random)
````

````python
plt.boxplot(com_random)
plt.show()
````
