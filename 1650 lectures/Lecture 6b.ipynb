### Plotting a Curve  
plot $y = 2 ^ x $ on [1, 5]

```python
xvals = np.linspace(1, 5, 100) # 100 equally spaced partition pts on [1, 5]
yvals = [2 ** x for x in xvals]

plt.plot(xvals, yvals)
plt.show()
```

Plot $y = \sin(x)$ on $[0, 2\pi]$

```python
xvals = np.linspace(0, 2 * math.pi, 100)
sinvals = [math.sin(x) for x in xvals]
cosvals = [math.cos(x) for x in xvals]
cosvalsneg = [- y for y in cosvals]

plt.plot(xvals, sinvals, ':', label = 'sin(x)')
plt.plot(xvals, cosvals, '--', label = 'cos(x)')
plt.plot(xvals, cosvalsneg, '-.', label = '-cos(x)')

plt.legend(loc = 'right')
plt.show()
```

```python
xvals = np.linspace(0, 2 * math.pi, 100)
cosvals = [math.cos(x) for x in xvals]
cosvals2 = [math.cos(x + 0.1) for x in xvals]
cosvals3 = [math.cos(x + 0.2) for x in xvals]

plt.plot(xvals, cosvals)
plt.plot(xvals, cosvals2)
plt.plot(xvals, cosvals3)
```

```python
xvals = np.linspace(0, 2 * math.pi, 100)

for i in range(20):
    cosvals = [math.cos(x + 1*i) for x in xvals]
    plt.plot(xvals, cosvals)
```
