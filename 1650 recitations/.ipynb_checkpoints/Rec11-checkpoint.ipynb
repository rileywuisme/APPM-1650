### Array Attributes

Write a function **`array_info(arr)`** that returns a tuple containing the following attributes of array `arr`:
 * number of dimensions
 * (number of rows, number of columns)
 * total number of elements
 * data type

For example, 
```
arr = np.arange(12).reshape(3, 4)
array_info(arr) 
```
returns
```
(2, (3, 4), 12, dtype('int64'))
```
or similar. (Note: The `dtype` might be slightly different for your computer.)



````python
def array_info(arr):
    dim = arr.ndim
    row_col = arr.shape
    tot = arr.size
    type = arr.dtype
    return (dim, row_col, tot, type)
````

````python
arr = np.arange(12).reshape(3, 4)
array_info(arr)
````

### Plotting with Vectorization
Plot the function $y = x\sin(10x)$ on $[0, \pi]$. **Use numpy vectorization** to calculate the function values.

````python
xval = np.linspace(0, np.pi, 100)
yval = xval * np.sin(10*xval)
plt.plot(xval, yval)
plt.show()
````

**Create a function `sinfunc(x)`** that returns the value of $x\sin(10x)$. Now use it to plot the function.

````python
def sinfunc(x):
    return x * np.sin(10*x)

xval = np.linspace(0, np.pi, 100)
plt.plot(xval, sinfunc(xval))
plt.show()
````

### Magic Squares

An $n\times n$ square filled with distinct positive integers is *semimagic* if its row sums and column sums are all equal. In a *magic* square, the diagonal sums also match the row and column sums. Here is an example of a magic square with a *magic sum* of 15:

|  |  |  |
|:--:|:--:|:--:|
|4|9|2|
|3|5|7|
|8|1|6|

Write functions **`is_semimagic(sq)`** and **`is_magic(sq)`** that take a two-dimensional numpy array representing an $n\times n$ square and return `True` if the square is semimagic or magic, respectively. The function `is_magic()` should call `is_semimagic()`. **Use numpy operations**.

Examples:<br>
`sq = np.arange(1, 10).reshape(3, 3)` is neither.<br>
`sqsemi = np.array([[3,7,5], [8,6,1], [4,2,9]])` is semimagic.<br>
`sqmagic = np.array([[4,9,2], [3,5,7], [8,1,6]])` is magic.

````python
sq = np.arange(1, 10).reshape(3, 3)
sqsemi = np.array([[3,7,5], [8,6,1], [4,2,9]])
sqmagic = np.array([[4,9,2], [3,5,7], [8,1,6]])
````

````python
def is_semimagic(sq):
    row_sum = sq.sum(axis=1)[0]
    col_sum = sq.sum(axis=0)[0]
    return row_sum == col_sum
````

````python
def is_magic(sq):
    row_sum = sq.sum(axis=1)[0]
    col_sum = sq.sum(axis=0)[0]
    for n in range(sq.shape[0]):
        diag = [sq[n, n]]
    return row_sum == col_sum == sum(diag)
````

````python
is_semimagic(sqsemi)
````

````python
is_magic(sqsemi)
````

### Franklin Squares
**Benjamin Franklin** enjoyed making magic squares. In 1771 he created a semimagic square, explaining that, <i>"I was at length tired with sitting there to hear debates, in which, as clerk, I could take no part, and which were often so unentertaining that I was induc'd to amuse myself with making magic squares or circles."</i> He later created a square which he described as <i>"the most magically magical of any magic square ever made by any magician"</i>.

Two of Franklin's squares are stored in `'fsquare1.txt'` and `'fsquare2.txt'`. They can be input as integer arrays by executing these commands:
```
fsquare1 = np.loadtxt('fsquare1.txt', dtype=int)
fsquare2 = np.loadtxt('fsquare2.txt', dtype=int)
```
Are these squares semimagic? magic? neither?


````python
fsquare1 = np.loadtxt('fsquare1.txt', dtype=int)
fsquare2 = np.loadtxt('fsquare2.txt', dtype=int)
````

````python
is_semimagic(fsquare1)
````

````python
is_magic(fsquare1)
````

````python
is_semimagic(fsquare2)
````

````python
is_magic(fsquare2)
````
