### A Sine Function
**Plot the function $y = x-3\sin x$ and its derivatives** $y'$ and $y''$ on $[-4,4]$ as line graphs.

<img src="http://www.coloradomath.org/python/sinegraphs1.jpg" width="432" height="288" />

````python
xvals = np.linspace(-4, 4, 100)

sinvals = [x - 3*math.sin(x) for x in xvals]
y_prime = [ -3 * math.cos(x) for x in xvals] 
y_double_prime = [3 * math.sin(x) for x in xvals]

plt.plot(xvals, sinvals)
plt.plot(xvals, y_prime)
plt.plot(xvals, y_double_prime)
plt.show()
````

### Exponential Functions
**Plot the curves** $y=a^x$ for $a=2,3,\ldots, 11$ and $a=\frac 12, \frac 13, \ldots, \frac{1}{11}$ on $[-5,5]$. Make the colors symmetric about the $y$-axis. Use `plt.ylim(ymin, ymax)` to set the range to $0\le y\le 50$.

<img src="http://www.coloradomath.org/python/exp-graphs.jpg" width="432" height="288" />

````python
xvals = np.linspace(-5, 5, 100)
for a in range(2, 12): 
    pos = [ a ** x for x in xvals]
    neg = [ ( 1/a ) ** x for x in xvals]
    color = 'C' + str(a-2)
    plt.plot(xvals, pos, color)
    plt.plot(xvals, neg, color)
    
plt.ylim(0, 50)
plt.show()
````

### A Circle
Draw a unit circle in two different ways: 
* by using the Cartesian equation: $y = f(x)$
* by using the parametric equations: $x=\cos(t)$, $y=\sin(t)$ 

(*Hint:* `plt.axis('equal')` will set the aspect ratio.)

<img src="http://www.coloradomath.org/python/circle.jpg" width="432" height="288" />

````python
x_values = np.linspace(-1, 1, 100)

y_upper = [math.sqrt(1 - x**2) for x in x_values]  
y_lower = [-math.sqrt(1 - x**2) for x in x_values] 

plt.plot(x_values, y_upper, 'b')
plt.plot(x_values, y_lower, 'b')

plt.axis('equal')
plt.show()
````

````python
t_values = np.linspace(0, 2 * math.pi, 100)

x_parametric = [math.cos(t) for t in t_values]
y_parametric = [math.sin(t) for t in t_values]

plt.plot(x_parametric, y_parametric, color='blue')

plt.axis('equal')

plt.show()
````

### Distribution of Random Numbers

Suppose we use `randint` to generate 10 random integers in the range from 1 to 10. The result might look like 
```
[3, 1, 7, 4, 5, 3, 10, 3, 5, 3].
```
Notice that the integer 1 appears once, the integer 2 does not appear, the integer 3 appears 4 times, etc. The list containing the frequencies of the integers from 1 to 10 is:
```
[1, 0, 4, 1, 2, 0, 1, 0, 0, 1].
```

Write a function **`random_freq(n)`** that generates `n` random integers from 1 to 10, then returns a list of the frequencies of the integers. Assume that `n` is a positive integer. (Do not use `count()`.)

**Display the result** of `random_freq(1000)` in a bar chart. Include horizontal grid lines by calling `plt.grid(axis='y')`. Here is a sample plot.

<img src="http://www.coloradomath.org/python/rand-freq.jpg" width="432" height="288" />

````python
def random_freq(n):
    random_10 = [random.randint(1, 10) for i in range(n)]
    freq = [0] * 10
    for number in random_10:
        freq[number - 1] += 1

    return freq

plt.bar(range(1, 11), random_freq(1000)) 
plt.grid(axis='y')
plt.show()
````
